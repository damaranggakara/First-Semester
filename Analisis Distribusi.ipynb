```python
# ======= NOMOR 1 =======
# DATA 1
jumlah_data = 200
mean_a = 80
simpangan_a = 5

# DATA 2
mean_b= 80
simpangan_b= 10

# PLOT DENSITY
data_a = np.random.normal(mean_a, simpangan_a, jumlah_data)
data_b = np.random.normal(mean_b, simpangan_b, jumlah_data)

# Visualize
sns.kdeplot(data_a, color='blue', label='N(80, 5)')
sns.kdeplot(data_b, color='red', label='N(80, 10)')
plt.title('Density Plot')
plt.legend()
plt.show()
```

```python
# INTERPRETASI NO 1
print(f"Plot di atas menunjukkan dua distribusi normal dengan rata-rata yang sama namun simpangan baku yang berbeda.\nDistribusi N(80,5) lebih terpusat dan kurang tersebar,\nsedangkan distribusi N(80,10) lebih datar dan tersebar luas.")
```

```python
# ========= NOMOR 2 ===========
# DATA BERDISTRIBUSI NORMAL
jumlah_data = 200
mean_2a = 50
simpangan_2a = 5

# DATA BERDISTRIBUSI BINOMIAL
mean_2b = 100
peluang_2b = 0.5

# PLOT DENSITY
data_c = np.random.normal(loc=mean_2a, scale=simpangan_2a, size=jumlah_data)
data_d = np.random.binomial(n=mean_2b, p=peluang_2b, size=jumlah_data)

# Visualize
sns.kdeplot(data_c, color='green', label='N(50, 5)')
sns.kdeplot(data_d, color='purple', label='Bin(100, 0.5, 0.5)')
plt.title('Density Plot')
plt.legend()
plt.show()
```

```python
# INTERPRETASI NO 2
print(f'Plot di atas membandingkan distribusi normal N(50,5) dengan distribusi binomial Bin(100,0.5).\nDistribusi normal (hijau) terlihat simetris dan mulus,\nsedangkan distribusi binomial(ungu) menunjukkan puncak di sekitar nilai 50 dengan variasi yang lebih lebar.')
```

```python
# ====== NOMOR 3 =======
poisson_data = np.random.poisson(50, 200)
poisson_mean = np.mean(poisson_data)
poisson_variance = np.var(poisson_data)

print('Poisson distribution mean:', poisson_mean)
print('Poisson distribution variance:', poisson_variance)

# INTERPRETASI
print(f"Rata-rata dan varians dari distribusi Poisson yang dihasilkan saling mendekati, \nyang merupakan karakteristik dari distribusi Poisson.\nRata-rata sedikit lebih rendah dari nilai yang diharapkan sebesar 50, dan varians sedikit lebih tinggi,\nmenunjukkan variasi yang tipikal akibat sifat acak dari proses penghasilan data.")
```
